# Raw strings

Você pode colocar um r antes das aspas iniciais de uma string para torná-la uma raw string, que ignora completamente todos os caracteres de escape e imprime qualquer barra invertida que apareça na string. 

```python
print(r'That is Carol\'s cat.')
```

# Colocando strings dentro de outras strings
Colocar strings dentro de outras strings é uma operação comum em programação. 

```python
name = 'Al'
age = 4000
'Hello, my name is ' + name + '. I am ' + str(age) + ' years old.'
```

```python
name = 'Al'
age = 4000
'My name is %s. I am %s years old.' % (name, age)
```

```python
name = 'Al'
age = 4000
f'My name is {name}. Next year I will be {age + 1}.'
```

Os métodos upper(), lower(), isupper() e islower()

```python
spam = 'Hello, world!'
spam = spam.upper()
spam
```

```python
spam = spam.lower()
spam
```

Os métodos upper() e lower() são úteis se você precisar fazer uma comparação sem distinção entre maiúsculas e minúsculas. Por exemplo, as strings 'great' e 'GREat' não são iguais entre si. Mas no pequeno programa a seguir, não importa se o usuário digita Great , GREAT ou grEAT , porque a string é primeiro convertida em minúsculas.

```python
print('How are you?')
feeling = input()
if feeling.lower() == 'great':
    print('I feel great too.')
else:
    print('I hope the rest of your day is good.')
```

```python
spam = 'Hello, world!'
spam.islower()
spam.isupper()
'HELLO'.isupper()
'abc12345'.islower()
'12345'.islower()
'12345'.isupper()
```
